##‚úÖQuestion 1: What is hypothesis testing in statistics?
##Answer:
Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.  
It helps us determine whether there is enough evidence to **reject a claim (null hypothesis)** about the population.

---

### Key Concepts:
1. **Null Hypothesis (H‚ÇÄ) :**
   - The default assumption or claim (e.g., "There is no difference", "The mean = 50").
   
2. **Alternative Hypothesis (H‚ÇÅ or Ha):**
   - The claim we want to test (e.g., "There is a difference", "The mean ‚â† 50").
   
3. **Test Statistic:**
   - A value calculated from sample data to compare against a critical value.
   
4. **P-value:**
   - Probability of getting a result at least as extreme as the observed one, assuming H‚ÇÄ is true.
   - If p-value ‚â§ significance level (Œ±, usually 0.05), we **reject H‚ÇÄ**.

5. **Types of Errors:**
   - **Type I Error (Œ±):** Rejecting H‚ÇÄ when it is true.
   - **Type II Error (Œ≤):** Failing to reject H‚ÇÄ when it is false.

---

### Steps in Hypothesis Testing:
1. Define the null (H‚ÇÄ) and alternative hypothesis (H‚ÇÅ).
2. Choose a significance level (Œ±).
3. Collect sample data and compute the test statistic.
4. Find the p-value or compare with the critical value.
5. Make a decision: **Reject H‚ÇÄ or Fail to Reject H‚ÇÄ**.

---

### Example in Python: One-sample t-test
import numpy as np

from scipy import stats

### Sample data: Exam scores of 10 students
data = [82, 85, 90, 88, 76, 95, 89, 84, 91, 87]

### Hypothesis: Mean score = 80
### H0: mean = 80
### H1: mean ‚â† 80

t_stat, p_value = stats.ttest_1samp(data, 80)

print("T-statistic:", t_stat)

print("P-value:", p_value)

 - if p_value < 0.05 :

    print("Reject the Null Hypothesis (H0).")

 - else:

    print("Fail to Reject the Null Hypothesis (H0).")


##‚úÖQuestion 2 : What is the null hypothesis, and how does it differ from the alternative hypothesis?
##Answer:
The **null hypothesis (H‚ÇÄ)** is the default assumption in statistics that there is **no effect, no difference, or no relationship** in the population.  
It represents the claim we start with before analyzing data.

The **alternative hypothesis (H‚ÇÅ or Ha)** is the statement that **contradicts the null hypothesis**.  
It represents what we are trying to prove ‚Äî that there is an effect, a difference, or a relationship.

---

### Key Differences:
| Aspect | Null Hypothesis (H‚ÇÄ) | Alternative Hypothesis (H‚ÇÅ) |
|--------|----------------------|-----------------------------|
| Definition | Assumes no effect or difference | Assumes an effect or difference exists |
| Example | Mean = 50 | Mean ‚â† 50 (two-tailed) or Mean > 50 / Mean < 50 (one-tailed) |
| Purpose | Starting assumption for testing | Competing claim to be tested |
| Decision | Either "Fail to reject H‚ÇÄ" or "Reject H‚ÇÄ" | Accepted when H‚ÇÄ is rejected |

---

### Types of Alternative Hypotheses:
1. **Two-tailed test:** H‚ÇÅ: Œº ‚â† Œº‚ÇÄ  
2. **Right-tailed test:** H‚ÇÅ: Œº > Œº‚ÇÄ  
3. **Left-tailed test:** H‚ÇÅ: Œº < Œº‚ÇÄ  

---

##‚úÖQuestion 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

##Answer:
The **significance level (Œ±)** is the threshold used in hypothesis testing to decide whether to **reject the null hypothesis (H‚ÇÄ)**.  
It represents the probability of making a **Type I error** (rejecting H‚ÇÄ when it is actually true).  

Common choices for Œ± are **0.05 (5%)**, **0.01 (1%)**, or **0.10 (10%)**.

---

### Role in Hypothesis Testing:
1. **Set the Significance Level (Œ±):** Before testing, we decide how much risk of error we are willing to accept.  
   - Example: Œ± = 0.05 means we accept a 5% chance of wrongly rejecting H‚ÇÄ.
   
2. **Compare P-value with Œ±:**  
   - If **p-value ‚â§ Œ±** ‚Üí Reject H‚ÇÄ (evidence supports H‚ÇÅ).  
   - If **p-value > Œ±** ‚Üí Fail to reject H‚ÇÄ (not enough evidence against H‚ÇÄ).

---

### Key Points:
- A smaller Œ± makes the test more **conservative** (harder to reject H‚ÇÄ).  
- A larger Œ± makes the test more **sensitive** (easier to reject H‚ÇÄ, but increases risk of error).  
- Œ± directly controls the probability of a **false positive result**.  

---

### Example in Python: Demonstrating significance level
import numpy as np
from scipy import stats

### Sample data: Exam scores of students
data = [78, 82, 85, 80, 79, 83, 88, 84, 81, 77]

### Hypothesis:
### H0: mean score = 80
### H1: mean score ‚â† 80

alpha = 0.05   # significance level

print("T-statistic:", t_stat)

print("P-value:", p_value)

print("Significance level (alpha):", alpha)

 - if p_value < alpha:

    print("Reject H0: Evidence suggests the mean is significantly different from 80.")

 - else:

    print("Fail to Reject H0: No significant evidence against the mean = 80.")

##‚úÖQuestion 4: What are Type I and Type II errors? Give examples of each.

##Answer:
In hypothesis testing, errors can occur because we make decisions based on **sample data**, not the entire population.

---

### 1. **Type I Error (False Positive)**
- Occurs when we **reject the null hypothesis (H‚ÇÄ)** even though it is actually true.  
- Probability of Type I error = **Œ± (significance level)**.  
- Example:  
  - A medical test shows a patient **has a disease** (reject H‚ÇÄ: ‚Äúno disease‚Äù) when in reality the patient is **healthy**.  

---

### 2. **Type II Error (False Negative)**
- Occurs when we **fail to reject the null hypothesis (H‚ÇÄ)** even though it is actually false.  
- Probability of Type II error = **Œ≤**.  
- Example:  
  - A medical test fails to detect a disease (accept H‚ÇÄ: ‚Äúno disease‚Äù) when in reality the patient **has the disease**.  

---

### Key Difference:
| Error Type | Decision | Reality | Example |
|------------|----------|---------|---------|
| Type I (False Positive) | Reject H‚ÇÄ | H‚ÇÄ is true | Healthy person diagnosed as sick |
| Type II (False Negative) | Fail to Reject H‚ÇÄ | H‚ÇÄ is false | Sick person diagnosed as healthy |

---

##‚úÖQuestion 5: What is the difference between a Z-test and a T-test? Explain when to use each.
##Answer:
Both **Z-test** and **T-test** are statistical tests used to compare sample data with a population or between samples.  
The key difference lies in the **sample size** and whether the **population standard deviation (œÉ)** is known.

---

### üîπ Z-test
- Used when the **population standard deviation (œÉ) is known**.  
- Works best for **large sample sizes (n > 30)**.  
- Based on the **standard normal distribution (Z-distribution)**.  
- Example: Testing if the mean height of 10,000 students = 170 cm (œÉ known).

---

### üîπ T-test
- Used when the **population standard deviation (œÉ) is unknown**.  
- Suitable for **small sample sizes (n ‚â§ 30)**.  
- Based on the **Student‚Äôs t-distribution**, which accounts for extra uncertainty.  
- Example: Testing if the mean exam score of 15 students = 75 (œÉ unknown).

---

### Key Differences:
| Feature | Z-test | T-test |
|---------|--------|--------|
| Population Standard Deviation | Known | Unknown |
| Sample Size | Large (n > 30) | Small (n ‚â§ 30) |
| Distribution | Standard Normal (Z) | Student‚Äôs t |
| Use Case | Quality control, large datasets | Small experiments, sample studies |

---

##‚úÖQuestion 7:  Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
**sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**

(Include your Python code and output in the code box below.)
##Answer:
We will test whether the sample dataset has a mean equal to **Œº‚ÇÄ = 50** using the **Z-test**.

---

### Hypotheses:
- **Null Hypothesis (H‚ÇÄ):** Œº = 50  (the sample mean equals 50)  
- **Alternative Hypothesis (H‚ÇÅ):** Œº ‚â† 50  (the sample mean is different from 50)  

---

### Z-test Formula:
\[
Z = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}}
\]

Where:  
- \(\bar{X}\) = sample mean  
- \(\mu_0\) = hypothesized population mean  
- \(\sigma\) = sample standard deviation (used here since œÉ is unknown)  
- \(n\) = sample size  

Decision Rule:  
- If **p-value ‚â§ 0.05** ‚Üí Reject H‚ÇÄ  
- If **p-value > 0.05** ‚Üí Fail to Reject H‚ÇÄ  

---



```python
import numpy as np
from scipy import stats

# Sample dataset
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Sample statistics
n = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # unbiased estimator

# Z-test statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Size (n):", n)
print("Sample Mean:", round(sample_mean, 4))
print("Sample Std Dev:", round(sample_std, 4))
print("Z-statistic:", round(z_stat, 4))
print("P-value:", round(p_value, 6))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("‚úÖ Reject H0: The sample mean is significantly different from 50.")
else:
    print("‚ùå Fail to Reject H0: No significant evidence that the mean differs from 50.")
```

##‚úÖQuestion 8:  Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)
##Answer:
We will:  
1. Generate random data from a **normal distribution**.  
2. Calculate the **95% confidence interval (CI)** for the population mean using the formula:

\[
CI = \bar{X} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
\]

Where:  
- \(\bar{X}\) = sample mean  
- \(Z_{\alpha/2}\) = 1.96 for 95% confidence  
- \(\sigma\) = sample standard deviation  
- \(n\) = sample size  

3. Plot the data distribution with **mean and CI bounds**.  

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate normal distribution data
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=200)  # mean=50, std=5, n=200

# Step 2: Calculate 95% confidence interval
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# Z critical value for 95% confidence
z_critical = stats.norm.ppf(0.975)  # two-tailed, 0.975
margin_of_error = z_critical * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("Sample Mean:", round(sample_mean, 3))
print("95% Confidence Interval:", (round(ci_lower, 3), round(ci_upper, 3)))

# Step 3: Plot histogram of the data
plt.hist(data, bins=20, color="skyblue", edgecolor="black", alpha=0.7)
plt.axvline(sample_mean, color="red", linestyle="dashed", linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color="green", linestyle="dotted", linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color="green", linestyle="dotted", linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")

plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

##‚úÖQuestion 9:  Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

(Include your Python code and output in the code box below.)
##Answer:
### üîπ What is a Z-score?
- A **Z-score** tells us how many **standard deviations** a data point is away from the mean.  
- Formula:

\[
Z = \frac{X - \mu}{\sigma}
\]

Where:  
- \(X\) = data value  
- \(\mu\) = mean of the dataset  
- \(\sigma\) = standard deviation of the dataset  

### üîπ Interpretation:
- **Z = 0** ‚Üí value is exactly at the mean  
- **Z = +1** ‚Üí value is 1 standard deviation above the mean  
- **Z = -1** ‚Üí value is 1 standard deviation below the mean  
- Values with |Z| > 3 are often considered **outliers**.  

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample std deviation
    z_scores = [(x - mean) / std for x in data]
    return np.array(z_scores)

# Sample dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=200)  # normal distribution

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print few results
print("First 10 Z-scores:", np.round(z_scores[:10], 3))

# Plot histogram of Z-scores
plt.hist(z_scores, bins=20, color="lightcoral", edgecolor="black", alpha=0.7)
plt.title("Histogram of Standardized Data (Z-scores)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color="blue", linestyle="dashed", linewidth=2, label="Mean (Z=0)")
plt.legend()
plt.show()
```
